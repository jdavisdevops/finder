```python
from finder import Finder
import pandas as pd
import sqlite3
import re
import requests
import pandas as pd

f = Finder(
    search_term="women owned business", city="Pasadena", state="CA", headless=False
)
df = pd.read_sql(
    "SELECT * FROM women_owned_business", con=sqlite3.connect(f.search_file_path)
)
df.women_owned.replace("False", "", inplace=True)
df.website.replace("None", "", inplace=True)
print(len(df))
df.columns
```

```python
cols = ["title", "category", "address", "women_owned", "number", "website", "link"]
w = df[cols]
```

```python
def find_emails(link):
    if "http" not in link:
        link = "http://" + link
    response = requests.get(link)
    list_emails = set()
    new_emails = re.findall(
        r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b", response.text, re.I
    )
    list_emails.update(new_emails)
    list_emails = list(list_emails)
    list_emails = [x for x in list_emails if "png" not in x]
    list_emails = [x for x in list_emails if "jpg" not in x]
    list_emails = [x for x in list_emails if "wix" not in x]
    list_emails = [x for x in list_emails if "sentry" not in x]
    return set(list_emails)


# w = w.head(20)
w["email"] = None
for idx, x in w.iterrows():
    try:
        list_emails = find_emails(x["website"])
        if len(list_emails) > 0:
            w.loc[w.index == idx, "email"] = list_emails
    except:
        continue
```
